```python
# Dependencies
from splinter import Browser
from bs4 import BeautifulSoup
import pandas as pd
import requests
import pymongo
from flask import Flask, render_template, redirect
from flask_pymongo import PyMongo
from webdriver_manager.chrome import ChromeDriverManager
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://redplanetscience.com/'
browser.visit(url)
html=browser.html
soup=BeautifulSoup(html,'html.parser')
```

```python
for x in range(10):
    # Parent div for article title and teaser paragraph
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    results = soup.find_all('div',class_='col-md-8')

    # scrape loop
    for result in results:

        # Definitions
        news_title = result.find(class_='content_title')
        cleaned_title = news_title.text.strip()
        news_p = result.find(class_='article_teaser_body')
        cleaned_p = news_p.text.strip()

        # Prints
        print('---------------------------------')
        print(cleaned_title)
        print(cleaned_p)
 
```

```python
hemispheres=soup.find('div',class_='container')
results=hemispheres.find_all('div',class_='item')
for x in range(1):
    # Definitions
    marshemispheres = url = 'https://marshemispheres.com/'
    browser.visit(url)
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    results = soup.find_all('div', class_='item')

    hemisphere_list = []
    img_url_list = []
    hemisphere_img_dict =[] 

    # scrape loop
    for result in results:

        # Find featured image and URL
        hem=result.find('div',class_='description')
        title=hem.h3.text
        a = result.find('a')
        href = a['href']
        img = a.find('img')
        src = img['src']
        alt = img['alt']

        # Getting rid of unwanted words in the statement
        word_list = ["Enhanced", "thumbnail", "  "]

        for word in word_list:
            hemisphere = alt.replace(word, "") 
             
        img_url = marshemispheres + src

        # Prints
        print('---------------------------------')
        # print('alt')
        print(f"{hemisphere}:")
        print(f"{img_url}")

        # Dictionary appends
        dict={'title':title,'image_url':img_url}
        hemisphere_img_dict.append(dict)
```
